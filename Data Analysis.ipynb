# Sensory

```python
%matplotlib notebook

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import csv
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


gps_data=[]
with open("my_project/controllers/my_controller/gps_hand.csv", "r", newline='') as gps_csvfile:
    gps_reader = csv.reader(gps_csvfile)
   
    for row in gps_reader:
        gps_data.append([float(value) for value in row[1:]])  # Pasar la columna del indice
        
# Separar en tres listas: x, y, z
x, y, z = zip(*gps_data)

x = list(x)
y = list(y)
z = list(z)


colors = np.random.rand(len(x), 3)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z, c=colors, marker='o', s=1)

ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

plt.show()
plt.savefig('point_distribution.jpg')
                
```

```python
%matplotlib notebook

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import csv

gps_data = []
with open("my_project/controllers/my_controller/gps_hand.csv", "r", newline='') as gps_csvfile:
    gps_reader = csv.reader(gps_csvfile)
    for row in gps_reader:
        gps_data.append([float(value) for value in row[1:]])

x, y, z = zip(*gps_data)

# Convertir a listas
x = list(x)
y = list(y)
z = list(z)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z, c='blue', marker='o', s=1)  

ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

plt.show()
plt.savefig('point_distribution.jpg')
```

# Motor

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("my_project/controllers/my_controller/motor_angles.csv", header=None, names=["Index", "ShoulderPitch", "ShoulderRoll", "ElbowYaw", "ElbowRoll"])

plt.figure(figsize=(12, 6))
for i in range(len(data)):
    plt.plot(["ShoulderPitch", "ShoulderRoll", "ElbowYaw", "ElbowRoll"], 
             data.iloc[i, 1:], alpha=0.1, color='blue')  

plt.title("Gráfico de Coordenadas Paralelas para Ángulos de Motores")
plt.ylabel("Radianes")
plt.xlabel("Motores")
plt.grid(alpha=0.3)
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
projected = pca.fit_transform(data.iloc[:, 1:])

plt.figure(figsize=(8, 6))
plt.scatter(projected[:, 0], projected[:, 1], alpha=0.5, c='blue')
plt.title("Proyección PCA de los Datos de Entrenamiento")
plt.xlabel("1,2")
plt.ylabel("3,4")
plt.grid(alpha=0.3)
plt.show()
```
